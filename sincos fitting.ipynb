```python
import matplotlib.pyplot as plt
import math
import pandas as pd
import seaborn as sns
import numpy as np
from scipy import optimize
```

```python
pi=math.pi
```

```python
x=np.array(df['x'])
y=np.array(df['y'])
def fitting(X,y0,A,B,x1,w1,x2,w2):
  
    return y0+A*(np.sin(pi*(X-x1)/w1))**2+B*(np.cos(pi*(X-x2)/w2)) **2
popt, pcov = optimize.curve_fit(fitting, x, y,maxfev=100000)
#perr=np.sqrt(np.diag(pcov))
#calculate r_square
#residuals = y- test_func(x, *popt)
#ss_res = np.sum(residuals**2)
#ss_tot = np.sum((y-np.mean(y))**2)
#r_squared = 1 - (ss_res / ss_tot)
print(popt)
#print(r_squared)
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(x, y,color='cadetblue',s=30)
plt.plot(x, fitting(x, *popt), 'r', label= 'fit:\ny0=%f,\nA=%f,\nB=%f,\nx1=%f,\nw1=%f,\nx2=%f,\nw2=%f' % tuple(popt), )
#plt.plot(x, test_func(x, *popt), 'r-', label='fit:\nn=%f,\nu=%f' % tuple(popt))
plt.legend(loc='best',fontsize=10)
#plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0), useMathText=True)
#plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0), useMathText=True)
#plt.autoscale(enable=True, axis='y') 
#plt.ylim(0, 0.0000025)
#plt.xlim(-1, 1)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
#plt.xlabel('Temperature (K)',fontsize=20)
#plt.ylabel('$\\rho$ (\u03A9-m)' ,fontsize=20)
plt.show()
```
